```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
from io import BytesIO
import xlsxwriter
import time
import datetime
```

```python
main_url = r"file:///G:/scraping_project/scraping_practice_site_single_file_html.html"
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(main_url)
```

```python
try:
    element_id = driver.find_element(By.ID, "id-input")
    look_up_button = driver.find_element(By.ID, "lookup-btn")
except NoSuchElementException as e:
    print(f"Element Couldn't be found {e}")
```

```python
# Load existing Excel with pandas just to check IDs (optional)
df = pd.read_excel("ID_data.xlsx")
df.head()
```

```python
file_name = datetime.datetime.now().strftime(r"%Y-%m-%d %H-%M-%S%p") + ".xlsx"
```

```python
file_name
```

```python
workbook = xlsxwriter.Workbook(file_name)
worksheet = workbook.add_worksheet()

cell_format = workbook.add_format()
cell_format.set_align("vcenter")

worksheet.write(0, 0, "ID", cell_format)
worksheet.write(0, 1, "Screenshot", cell_format)
```

```python
# Loop through IDs and write collected data
for row, id_value in enumerate(df["id"], start=1):  # skips header or filled data
    # Clear field
    element_id.clear()
    # Fetching data
    element_id.send_keys(id_value)
    time.sleep(1)
    look_up_button.click()

    time.sleep(2)
    result = driver.find_element(By.ID, "result").screenshot_as_png

    image_data = BytesIO(result)

    worksheet.write(row, 0, f"{id_value}", cell_format)

    print(
        worksheet.embed_image(
            row,
            1,
            source=f"{id_value}.png",
            options={
                "image_data": image_data,
                "x_scale": 0.3,
                "y_scale": 0.3,
                "object_position": 1,
                "cell_format": cell_format,
            },
        )
    )

    time.sleep(3)

workbook.close()
```

```python
driver.quit()
```
